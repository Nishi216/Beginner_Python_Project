## Importing all the packages needed for creating the word cloud

```python
import wordcloud
from matplotlib import pyplot as plt
```

## Creating function to count frequency of words after cleaning it

```python
def calculate_frequencies(file_contents):
    # Here is a list of punctuations and uninteresting words you can use to process your text
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just"]
    
    new_string = ''
    # LEARNER CODE START HERE
    for chars in file_contents:
        if chars not in punctuations:
            new_string = new_string + chars.upper()
    
    word_list = []
    for words in new_string.split():
        if words in uninteresting_words:
            continue
        word_list.append(words)
    
    frequency_words = {}
    for word in word_list:
        if word in frequency_words:
            frequency_words[word] += 1
        else:
            frequency_words[word] = 1
    
    #wordcloud
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(frequency_words)
    cloud.to_file('names.png')
    return cloud.to_array()
```

## Displaying the word cloud image

```python
f = open('names.txt','r')
file_contents = f.readlines()
```

```python
myimage = calculate_frequencies(file_contents)
plt.imshow(myimage, interpolation = 'nearest')
plt.axis('off')
plt.show();
```
